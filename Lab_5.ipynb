Write a python program to perform **selection sort**.
- ** -Algorithm
* Step 1: Read the number of elements for the list from the user.
* Step 2: Using for loop insert the elements in the list.
* Step 3: Initialize the minimum element as min=numbers[i].
* Step 4: Using the swap method the elements are sorted accordingly.

```python
#write code for above cell and compute the running time and also plot graph.
import time
import random
import matplotlib.pyplot as plt


def selection_sort(numbers):
    n = len(numbers)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if numbers[j] < numbers[min_index]:
                min_index = j

        numbers[i], numbers[min_index] = numbers[min_index], numbers[i]
    return numbers



n = int(input("Enter the number of elements: "))

print(f"Enter {n} numbers separated by spaces:")
numbers_input_str = input()
numbers = [int(x) for x in numbers_input_str.split()]


selection_sort(numbers)
print("Sorted list:", numbers)



sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
times = []

for size in sizes:
    data = random.sample(range(1, 10000), size)

    start_time = time.time()
    selection_sort(data)
    end_time = time.time()

    times.append(end_time - start_time)


plt.plot(sizes, times, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (seconds)")
plt.title("Selection Sort Time Complexity")
plt.grid(True)
plt.show()
```

In computer science, **merge sort** (also commonly spelled mergesort) is an efficient, general-purpose, comparison-based sorting algorithm. Most implementations produce a stable sort, which means that the order of equal elements is the same in the input and output. Merge sort is a divide and conquer algorithm that was invented by John von Neumann in 1945.
Write a Python program to sort the elements using **merge sort** and plot a graph to the time taken versus n
- Algorithm(Merge Sort is a divide and conquer algorithm. It divides	input array in two halve.s, calls itself for the two halves and then merges two sorted halves.)
1.	Divide the unsorted array into n partitions, each	partition contains 1 element.
2.	Repeatedly merge partitioned units to produce	new sublists until there is only l sublist remaining. This will be the sorted	.
3.	Compare the first element of the sublist with the first element of the sublist to its right.
4.	Merge the two sublists by	paring each element of the sublist and placing the smaller element into the	new sublist.
5.	Repeat step 3 and 4 untill	all sublists are merged into a single sorted sublist.



```python
#write code for above cell and compute the running time and also plot graph.
import time
import random
import matplotlib.pyplot as plt

# -------- Merge Function --------
def merge(left, right):
    merged = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Add remaining elements
    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged


# -------- Merge Sort Function --------
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


# -------- User Input Section --------
n = int(input("Enter the number of elements: "))
# Fix: Prompt for all numbers on one line and parse them
print(f"Enter {n} numbers separated by spaces:")
numbers_input_str = input()
numbers = [int(x) for x in numbers_input_str.split()]

sorted_list = merge_sort(numbers)
print("Sorted list:", sorted_list)


# -------- Running Time Computation --------
sizes = [100, 200, 400, 800, 1600, 3200]
times = []

for size in sizes:
    data = random.sample(range(1, 100000), size)

    start = time.time()
    merge_sort(data)
    end = time.time()

    times.append(end - start)

# -------- Plot Graph --------
plt.plot(sizes, times, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (seconds)")
plt.title("Merge Sort Time Complexity")
plt.grid(True)
plt.show()
```

**Quicksort** (sometimes called partition-exchange sort) is an efficient sorting algorithm. Developed by British computer scientist Tony Hoare in 1959 and published in 1961, it is still a commonly used algorithm for sorting.

Quicksort is a divide-and-conquer algorithm. It works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot. The sub-arrays are then sorted recursively. This can be done in-place, requiring small additional amounts of memory to perform the sorting.
Write a Python program to sort the elements using quick sort and plot a graph to the time taken versus n.
- Algorithm
1.	Select a pivot element from the array.
2.	Partition the array into two sub-arrays. The elements in the first sub-array are less than the pivot element, while the elements in the second sub-array are greater than the pivot element.
3.	Recursively sort the sub-arrays created in Step 2.
4.	Join the sub-arrays and the pivot element together to	the sorted array.


```python
#write code for above cell and compute the running time and also plot graph.
import time
import random
import matplotlib.pyplot as plt

# -------- Partition Function --------
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1


# -------- Quick Sort Function --------
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)


# -------- User Input (Space Separated) --------
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

quick_sort(numbers, 0, len(numbers) - 1)
print("Sorted list:", numbers)


# -------- Running Time Computation --------
sizes = [100, 200, 400, 800, 1600, 3200]
times = []

for size in sizes:
    data = random.sample(range(1, 100000), size)

    start = time.time()
    quick_sort(data, 0, size - 1)
    end = time.time()

    times.append(end - start)

# -------- Plot Graph --------
plt.plot(sizes, times, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (seconds)")
plt.title("Quick Sort Time Complexity")
plt.grid(True)
plt.show()
```

```python
def selection_sort(arr):

    """
    Sorts a list in-place using the Selection Sort algorithm.
    Args:
        arr (list): The list of elements to be sorted.
    """
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        # Swap after finding minimum
        arr[i], arr[min_index] = arr[min_index], arr[i]

    print("Selection sort function defined.")

# You can test the function with a sample list:
# test_list = [64, 25, 12, 22, 11]
# selection_sort(test_list)
# print("Sorted test list:", test_list)
```

## Implement Merge Sort

### Subtask:
Write the Python code for the `merge_sort` algorithm in cell `okJbG3cRoITp` based on the provided algorithm description. This function should sort a list in-place.


**Reasoning**:
The subtask requires implementing the `merge_sort` algorithm as a Python function, following the detailed steps for recursive division and merging. I will create a `code_block` for cell `okJbG3cRoITp` containing this function and a print statement to confirm its definition.



```python
def merge_sort(arr):
    """
    Sorts a list in-place using the Merge Sort algorithm.
    Args:
        arr (list): The list of elements to be sorted.
    """
    if len(arr) > 1:
        mid = len(arr) // 2  # Find the middle of the list
        left_half = arr[:mid]  # Divide the elements into two halves
        right_half = arr[mid:]

        merge_sort(left_half)  # Sort the first half
        merge_sort(right_half)  # Sort the second half

        i = 0  # Initial index for left_half
        j = 0  # Initial index for right_half
        k = 0  # Initial index for merged list arr

        # Merge the two halves
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Copy remaining elements of left_half, if any
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        # Copy remaining elements of right_half, if any
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

print("Merge sort function defined.")

# You can test the function with a sample list:
# test_list = [38, 27, 43, 3, 9, 82, 10]
# merge_sort(test_list)
# print("Sorted test list:", test_list)
```

## Implement Quick Sort

### Subtask:
Write the Python code for the `quick_sort` algorithm in cell `HvucalAUoITq` based on the provided algorithm description. This function should sort a list in-place.


**Reasoning**:
The subtask requires implementing the `quick_sort` algorithm as a Python function, including helper functions for partitioning and recursive calls, and then adding a print statement to confirm its definition.



```python
def quick_sort(arr):
    """
    Sorts a list in-place using the Quick Sort algorithm.
    Args:
        arr (list): The list of elements to be sorted.
    """
    def _quick_sort_helper(arr, low, high):
        if low < high:
            # pi is partitioning index, arr[p] is now at right place
            pi = partition(arr, low, high)

            # Separately sort elements before partition and after partition
            _quick_sort_helper(arr, low, pi - 1)
            _quick_sort_helper(arr, pi + 1, high)

    def partition(arr, low, high):
        pivot = arr[high]  # Choose the last element as the pivot
        i = low - 1  # Index of smaller element

        for j in range(low, high):
            # If current element is smaller than or equal to pivot
            if arr[j] <= pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i] # Swap it with the element at i

        arr[i + 1], arr[high] = arr[high], arr[i + 1] # Swap pivot with the element at i+1
        return i + 1

    _quick_sort_helper(arr, 0, len(arr) - 1)

print("Quick sort function defined.")

# You can test the function with a sample list:
# test_list = [10, 7, 8, 9, 1, 5]
# quick_sort(test_list)
# print("Sorted test list:", test_list)
```
